#**Corelation Analysis**

```python
import pandas as pd

# Define transformation function
def split_features(value):
    if 'BHK' in value:
        # Extract the number before 'BHK' for bedroom
        bedroom = int(value.split()[0])
        hall = 1
        kitchen = 1
    elif 'Bedroom' in value:
        # Extract the number before 'Bedroom' for bedroom
        bedroom = int(value.split()[0])
        hall = 0
        kitchen = 0
    elif value == '1 RK':
        bedroom = 1
        hall = 0
        kitchen = 0
    else:
        bedroom = None
        hall = None
        kitchen = None
    return bedroom, hall, kitchen

# Apply the function to create new columns
df[['Bedroom', 'Hall', 'Kitchen']] = df['size'].apply(lambda x: pd.Series(split_features(x)))

# Display the DataFrame
df
```

```python
df.drop('size',axis=1,inplace=True)
```

```python
df['area_type'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
df['location']= encoder.fit_transform(df['location'])
df['availability']= encoder.fit_transform(df['availability'])
```

```python
# Replacing area type because it is ordinal data

df['area_type'].replace(['Plot  Area','Carpet  Area','Built-up  Area','Super built-up  Area'],[1,2,3,4],inplace=True)
df
```

```python
df['total_sqft'].unique() #these are string values. we cant give astype int because it has 1133-1384 range value. so we need to convert this data.
```

```python
import numpy as np

def convert_range_to_num(x):
    try:
        tokens = x.split('-')
        if len(tokens) == 2:
            return (float(tokens[0]) + float(tokens[1])) / 2
        try:
            return float(x)
        except ValueError:
            return np.nan  # Handle cases with units like 'Sq. Meter' or 'Sq. Yards'
    except:
        return np.nan

# Apply the function to the 'total_sqft' column
df['total_sqft'] = df['total_sqft'].apply(convert_range_to_num)
```

#**Removing Outlayers**

```python
df.boxplot()
df.shape
```

```python
def removeoutlier(df,exclude_col):
  for col in df.columns:
    if col == exclude_col:
      continue
    q1 = df[col].quantile(0.25)
    q3= df[col].quantile(0.75)
    IQR = q3-q1
    lower_bound =q1 - 1.5 * IQR
    upper_bound = q3 + 1.5 * IQR
    df = df[(df[col]>=lower_bound)& (df[col]<= upper_bound)]
  return df

df_cleaned = removeoutlier(df,exclude_col='price')
df = df_cleaned
```

```python
df
```

```python
x = df.drop('price',axis=1)
y = df['price']

xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2)
model = LinearRegression()
model.fit(xtrain,ytrain)
ypred = model.predict(xtest)
print(ypred)
```

```python
accuracy = model.score(xtest,ytest)
print(accuracy)
```

#**Testing in different models**

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import train_test_split

# Assuming 'target_column' is the name of the column you want to predict
x = df.drop(['price'], axis=1)
y = df['price']
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)

models = {
    'Linear Regression': LinearRegression(),
    'Decision Tree': DecisionTreeRegressor(),
    'Random Forest': RandomForestRegressor(),
    'SVR': SVR(),
    'KNN': KNeighborsRegressor()
}

for model_name, model in models.items():
    model.fit(xtrain, ytrain)
    ypred = model.predict(xtest)
    mse = mean_squared_error(ytest, ypred)
    r2 = r2_score(ytest, ypred)
    print(f'{model_name}')
    print(f'Mean Squared Error: {mse:.2f}')
    print(f'R2 Score: {r2:.2f}')
    print('_' * 30)
```

```python
model = KNeighborsRegressor()
model.fit(xtrain,ytrain)
ypred = model.predict(xtest)
mse = mean_squared_error(ytest, ypred)
r2 = r2_score(ytest, ypred)
print(ypred)
print(mse)
print(r2)
```
