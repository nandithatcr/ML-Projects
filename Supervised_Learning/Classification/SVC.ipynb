Accuracy Without Scaling

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

x = df.drop('survived',axis = 1)
y = df['survived']
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2)
model = SVC()
model.fit(xtrain,ytrain)
ypred = model.predict(xtest)
print(ypred)
```

```python
accuracy = model.score(xtest,ytest)
accuracy
```

```python
from sklearn.preprocessing import StandardScaler

features = ['pclass', 'sex', 'fare', 'alone']
scaler = StandardScaler()

std_scaled_data  = scaler.fit_transform(df[features])
std_scaled_df = pd.DataFrame(std_scaled_data,columns = features, index = df.index)
std_scaled_df
```

```python
x = std_scaled_df[['pclass', 'sex', 'fare', 'alone']]
y = df['survived']
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2)
std_model = SVC()
std_model.fit(xtrain,ytrain)
ypred = std_model.predict(xtest)
ypred
```

```python
accuracy = std_model.score(xtest,ytest)
accuracy
```

```python
from sklearn.preprocessing import MinMaxScaler
features = ['pclass', 'sex', 'fare', 'alone']
scaler = MinMaxScaler()
min_scaled_data  = scaler.fit_transform(df[features])

min_scaled_df = pd.DataFrame(min_scaled_data,columns = features, index = df.index)
min_scaled_df
```

```python
x = min_scaled_df[['pclass', 'sex', 'fare', 'alone']]
y = df['survived']
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2)
model = SVC()
model.fit(xtrain,ytrain)
ypred = model.predict(xtest)
ypred
```

```python
accuracy = model.score(xtest,ytest)
accuracy
```

Compare the accuracy using different Kernels

[Refer naive bayes algorithm and apriori algorithm]

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

x = min_scaled_df[['pclass', 'sex', 'fare', 'alone']]
y = df['survived']
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.2,random_state= 42)

# sv_linear = SVC(kernel="linear")
# sv_linear.fit(xtrain, ytrain)
# sv_linear.score(xtest, ytest)


models={'sv_linear': SVC(kernel="linear"),
        'sv_poly': SVC(kernel='poly', degree=8),
        'sv_gaussian': SVC(kernel ='rbf', random_state = 42),
        'sv_sigmoid': SVC(kernel ='sigmoid')
        }

for model_name,model in models.items():
  model.fit(xtrain,ytrain)
  ypred = model.predict(xtest)
  accuracy = accuracy_score(ytest,ypred)
  print(f'{model_name} accuracy: {accuracy:.2f}')
  print('Classification Report: ')
  print(classification_report(ytest,ypred))
  print('_'*30)
```
