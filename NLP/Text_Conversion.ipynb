```python
from googletrans import Translator
from gtts import gTTS
from IPython.display import Audio

# Initialize the translator
translator = Translator()

# English text to be translated
english_text ='''Go there'''

# Translate to Malayalam
translated_text = translator.translate(english_text, src='en', dest='ml').text
print(f"Translated Text: {translated_text}")

# Convert Malayalam text to speech
tts = gTTS(translated_text, lang='ml')

# Save the audio file
tts.save("malayalam_audio.mp3")

# Play t
# Play the audio in Colab
Audio("malayalam_audio.mp3", autoplay=True)
```

#Convert to Image

```python
from diffusers import StableDiffusionPipeline
import torch

# Load the Stable Diffusion model
pipe = StableDiffusionPipeline.from_pretrained("runwayml/stable-diffusion-v1-5")
pipe = pipe.to("cuda")  # Use GPU for faster inference

# Text input
text = "christiano ronaldo"

# Generate an image
image = pipe(text).images[0]

# Save and display the image
output_image_path = "generated_image.png"
image.save(output_image_path)
print(f"Image saved as {output_image_path}")

# Optionally display the image
image.show()
```
